## Simplified aluminium production system

In this group exercise, we model a simplified life cycle assessment system for the production of 1 kg of primary aluminium at the smelter gate. The purpose of the case study is not to represent a complete real-world aluminium supply chain, but to practice the Brightway modelling logic introduced in the course: building a supply chain as a graph made of nodes and edges, defining products and processes, adding biosphere flows, and comparing scenarios.

The foreground system is based on a simple aluminium production chain:

- bauxite mining,
- alumina production,
- electricity production in a power plant,
- aluminium smelting,
- final product: 1 kg aluminium.

The functional unit is defined as **1 kg primary aluminium at the smelter gate**.

In Brightway terms, the system is represented as a graph. The process nodes are bauxite mining, alumina production, power plant, and aluminium smelter. The product nodes are bauxite, alumina, electricity, and aluminium. The edges between product and process nodes represent technosphere exchanges. For example, the aluminium smelter consumes 1.9 kg of alumina and 15 kWh of electricity to produce 1 kg of aluminium. The bauxite mining process supplies 4.6 kg of bauxite to alumina production.

The only biosphere flow considered in this simplified model is **CO₂ emitted to air**. Therefore, all environmental interventions are represented as CO₂ biosphere exchanges. The assumed direct CO₂ emissions are:

- bauxite mining: 0.1 kg CO₂,
- alumina production: 2.9 kg CO₂,
- aluminium smelter: 2.7 kg CO₂.

The main scenario difference is the electricity mix used by the power plant. The aluminium value chain itself is kept the same in all scenarios, but the CO₂ emissions from electricity production change depending on the assumed regional electricity mix. This allows us to compare how the same aluminium system changes when produced using different electricity supply assumptions.

We define three simplified electricity scenarios:

1. **Norway scenario**  
   This scenario represents a relatively low-carbon electricity system. We assume 50% hydropower, 30% wind power, and 20% solar power. Hydropower and wind are given low CO₂ emission factors, while solar is assumed slightly higher because of upstream production of panels. The weighted emission factor is 0.023 kg CO₂/kWh. For 15 kWh electricity, this gives 0.35 kg CO₂.

2. **Middle East scenario**  
   This scenario represents a fossil-based electricity system. We assume 70% natural gas, 20% oil, and 10% solar. Natural gas is assumed lower than oil, but both are much higher than renewable electricity. The weighted emission factor is 0.47 kg CO₂/kWh. For 15 kWh electricity, this gives 7.05 kg CO₂.

3. **China scenario**  
   This scenario represents a mixed but coal-dominated electricity system. We assume 60% coal, 20% natural gas, 10% hydropower, and 10% wind/solar. Coal has the highest assumed CO₂ emission factor, so this scenario has the highest electricity-related emissions. The weighted emission factor is 0.635 kg CO₂/kWh. For 15 kWh electricity, this gives 9.53 kg CO₂.

The total CO₂ impact for each scenario is calculated by adding the fixed process emissions from bauxite mining, alumina production, and aluminium smelting to the scenario-specific electricity emissions.

Fixed emissions:

0.1 + 2.9 + 2.7 = 5.7 kg CO₂

Scenario totals:

- Norway: 5.7 + 0.35 = 6.05 kg CO₂
- Middle East: 5.7 + 7.05 = 12.75 kg CO₂
- China: 5.7 + 9.53 = 15.23 kg CO₂

This case study helps us understand how Brightway represents LCA systems internally. The model is not only a calculation, but a structured graph consisting of process nodes, product nodes, technosphere edges, production edges, and biosphere edges. By changing only the electricity process while keeping the rest of the graph constant, we can compare regional production scenarios in a transparent way.
